```python
import random

def ruina_jugador(capital_inicial=3, prob_ganar=0.6, n=4):
    capital = capital_inicial
    historial = [capital]  # Guardamos el camino
    
    for i in range(n):
        if capital == 0 or capital == 8:  # estados absorbentes
            break
        if random.random() < prob_ganar:  # gana $1
            capital += 1
        else:  # pierde $1
            capital -= 1
        historial.append(capital)
    
    return historial

# Simulamos varias veces
for i in range(5):
    print(f"Simulación {i+1}: {ruina_jugador()}")
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Configuración
fig, ax = plt.subplots(figsize=(16, 12))

# Función para dibujar un nodo circular
def dibujar_nodo(ax, x, y, fortuna, paso, color, es_especial=None):
    circle = plt.Circle((x, y), 0.35, color=color, ec='black', lw=2.5, zorder=3)
    ax.add_patch(circle)
    ax.text(x, y, f'${fortuna}', ha='center', va='center', 
            fontsize=11, fontweight='bold', color='white', zorder=4)
    ax.text(x, y-0.65, f'n={paso}', ha='center', va='center', 
            fontsize=8, color='gray', zorder=4)
    
    # Marcar nodos especiales
    if es_especial == 'meta':
        ax.text(x, y+0.65, '★ META', ha='center', fontsize=7, 
                color='gold', fontweight='bold', zorder=4)
    elif es_especial == 'ruina':
        ax.text(x, y+0.65, '☠ RUINA', ha='center', fontsize=7, 
                color='red', fontweight='bold', zorder=4)

# Función para dibujar una línea con etiqueta
def dibujar_linea(ax, x1, y1, x2, y2, etiqueta, color_linea='black'):
    ax.plot([x1, x2], [y1, y2], color=color_linea, lw=1.8, zorder=1, alpha=0.7)
    mid_x, mid_y = (x1 + x2) / 2, (y1 + y2) / 2
    ax.text(mid_x-0.15, mid_y + 0.2, etiqueta, ha='center', va='bottom',
            fontsize=7, bbox=dict(boxstyle='round,pad=0.2', facecolor='white', 
            edgecolor='none', alpha=0.9), zorder=2)

# Colores por nivel
colores = {
    0: '#10b981',  # Verde
    1: '#ef4444',  # Rojo
    2: '#06b6d4',  # Cyan
    3: '#8b5cf6',  # Púrpura
    4: '#f59e0b'   # Naranja
}

# Definir todas las posiciones del árbol completo hasta n=4
# Estructura: {(fortuna, paso): (x, y)}

# n=0: solo $3
pos = {(3, 0): (2, 8)}

# n=1: $4 y $2
pos.update({
    (4, 1): (4, 10),
    (2, 1): (4, 6)
})

# n=2: $5, $3 (dos veces), $1
pos.update({
    (5, 2): (6, 12),
    (3, 2): (6, 9),      # viene de 4
    (3, 2.1): (6, 7),    # viene de 2 (segundo nodo $3)
    (1, 2): (6, 4)
})

# n=3: $6, $4 (dos veces), $2 (dos veces), $0
pos.update({
    (6, 3): (8, 13.5),
    (4, 3): (8, 11.5),    # viene de 5
    (4, 3.1): (8, 10),    # viene de 3 (primer)
    (2, 3): (8, 8),       # viene de 3 (primer)
    (2, 3.1): (8, 6),     # viene de 3 (segundo)
    (4, 3.2): (8, 5),     # viene de 3 (segundo)
    (2, 3.2): (8, 3.5),   # viene de 1
    (0, 3): (8, 2)
})

# n=4: $7, $5 (múltiples), $3 (múltiples), $1 (múltiples)
y_n4 = [14.5, 13.5, 12.5, 11.5, 10.5, 9.5, 8.5, 7.5, 6.5, 5.5, 4.5, 3.5, 2.5, 1.5, 0.5]
pos_n4 = [
    (7, 4, 14.5),  # de 6
    (5, 4, 13.5),  # de 6
    (5, 4.1, 12.5),  # de 4 (primero)
    (3, 4, 11.5),  # de 4 (primero)
    (5, 4.2, 10.5),  # de 4 (segundo)
    (3, 4.1, 9.5),  # de 4 (segundo)
    (3, 4.2, 8.5),  # de 2 (primero)
    (1, 4, 7.5),   # de 2 (primero)
    (3, 4.3, 6.5),  # de 2 (segundo)
    (1, 4.1, 5.5),  # de 2 (segundo)
    (5, 4.3, 4.5),  # de 4 (tercero)
    (3, 4.4, 3.5),  # de 4 (tercero)
    (3, 4.5, 2.5),  # de 2 (tercero)
    (1, 4.2, 1.5)   # de 2 (tercero)
]

for fortuna, paso, y in pos_n4:
    pos[(fortuna, paso)] = (10, y)

# DIBUJAR N=0
dibujar_nodo(ax, *pos[(3,0)], 3, 0, colores[0])

# DIBUJAR N=1
dibujar_linea(ax, *pos[(3,0)], *pos[(4,1)], 'p=0.6', 'green')
dibujar_linea(ax, *pos[(3,0)], *pos[(2,1)], 'q=0.4', 'red')
dibujar_nodo(ax, *pos[(4,1)], 4, 1, colores[1])
dibujar_nodo(ax, *pos[(2,1)], 2, 1, colores[1])

# DIBUJAR N=2
dibujar_linea(ax, *pos[(4,1)], *pos[(5,2)], 'p=0.6', 'green')
dibujar_linea(ax, *pos[(4,1)], *pos[(3,2)], 'q=0.4', 'red')
dibujar_linea(ax, *pos[(2,1)], *pos[(3,2.1)], 'p=0.6', 'green')
dibujar_linea(ax, *pos[(2,1)], *pos[(1,2)], 'q=0.4', 'red')

dibujar_nodo(ax, *pos[(5,2)], 5, 2, colores[2])
dibujar_nodo(ax, *pos[(3,2)], 3, 2, colores[2])
dibujar_nodo(ax, *pos[(3,2.1)], 3, 2, colores[2])
dibujar_nodo(ax, *pos[(1,2)], 1, 2, colores[2])

# DIBUJAR N=3
dibujar_linea(ax, *pos[(5,2)], *pos[(6,3)], 'p=0.6', 'green')
dibujar_linea(ax, *pos[(5,2)], *pos[(4,3)], 'q=0.4', 'red')
dibujar_linea(ax, *pos[(3,2)], *pos[(4,3.1)], 'p=0.6', 'green')
dibujar_linea(ax, *pos[(3,2)], *pos[(2,3)], 'q=0.4', 'red')
dibujar_linea(ax, *pos[(3,2.1)], *pos[(4,3.2)], 'p=0.6', 'green')
dibujar_linea(ax, *pos[(3,2.1)], *pos[(2,3.1)], 'q=0.4', 'red')
dibujar_linea(ax, *pos[(1,2)], *pos[(2,3.2)], 'p=0.6', 'green')
dibujar_linea(ax, *pos[(1,2)], *pos[(0,3)], 'q=0.4', 'red')

dibujar_nodo(ax, *pos[(6,3)], 6, 3, colores[3])
dibujar_nodo(ax, *pos[(4,3)], 4, 3, colores[3])
dibujar_nodo(ax, *pos[(4,3.1)], 4, 3, colores[3])
dibujar_nodo(ax, *pos[(2,3)], 2, 3, colores[3])
dibujar_nodo(ax, *pos[(2,3.1)], 2, 3, colores[3])
dibujar_nodo(ax, *pos[(4,3.2)], 4, 3, colores[3])
dibujar_nodo(ax, *pos[(2,3.2)], 2, 3, colores[3])
dibujar_nodo(ax, *pos[(0,3)], 0, 3, colores[3], 'ruina')

# DIBUJAR N=4
conexiones_n4 = [
    ((6,3), (7,4), 'green'),
    ((6,3), (5,4), 'red'),
    ((4,3), (5,4.1), 'green'),
    ((4,3), (3,4), 'red'),
    ((4,3.1), (5,4.2), 'green'),
    ((4,3.1), (3,4.1), 'red'),
    ((2,3), (3,4.2), 'green'),
    ((2,3), (1,4), 'red'),
    ((2,3.1), (3,4.3), 'green'),
    ((2,3.1), (1,4.1), 'red'),
    ((4,3.2), (5,4.3), 'green'),
    ((4,3.2), (3,4.4), 'red'),
    ((2,3.2), (3,4.5), 'green'),
    ((2,3.2), (1,4.2), 'red')
]

for origen, destino, color in conexiones_n4:
    etiqueta = 'p=0.6' if color == 'green' else 'q=0.4'
    dibujar_linea(ax, *pos[origen], *pos[destino], etiqueta, color)

# Dibujar nodos n=4
for fortuna, paso, _ in pos_n4:
    es_especial = 'meta' if fortuna >= 8 else ('ruina' if fortuna == 0 else None)
    dibujar_nodo(ax, *pos[(fortuna, paso)], fortuna, 4, colores[4], es_especial)

# Título y leyenda
ax.text(6, 16, 'Árbol Completo de Fortuna del Jugador (n=0 hasta n=4)', 
        ha='center', fontsize=16, fontweight='bold')
ax.text(6, 15.3, 'Fortuna Inicial: $3  |  p=0.6 (ganar +$1)  |  q=0.4 (perder -$1)', 
        ha='center', fontsize=10)

# Leyenda
leyenda = [
    mpatches.Patch(color=colores[0], label='n=0'),
    mpatches.Patch(color=colores[1], label='n=1'),
    mpatches.Patch(color=colores[2], label='n=2'),
    mpatches.Patch(color=colores[3], label='n=3'),
    mpatches.Patch(color=colores[4], label='n=4'),
]
ax.legend(handles=leyenda, loc='upper left', fontsize=9)

# Configuración
ax.set_xlim(0, 12)
ax.set_ylim(0, 17)
ax.axis('off')

plt.tight_layout()
plt.show()

# Calcular probabilidades
print("\n" + "="*70)
print("ÁRBOL COMPLETO: TODOS LOS CAMINOS DE n=0 A n=4")
print("="*70)

caminos = [
    ([3,4,5,6,7], 0.6**4, "→ $8 META en siguiente jugada"),
    ([3,4,5,6,5], 0.6**3*0.4, ""),
    ([3,4,5,4,5], 0.6**2*0.4*0.6, ""),
    ([3,4,5,4,3], 0.6**2*0.4**2, ""),
    ([3,4,3,4,5], 0.6*0.4*0.6**2, ""),
    ([3,4,3,4,3], 0.6*0.4*0.6*0.4, ""),
    ([3,4,3,2,3], 0.6*0.4**2*0.6, ""),
    ([3,4,3,2,1], 0.6*0.4**3, ""),
    ([3,2,3,4,5], 0.4*0.6**3, ""),
    ([3,2,3,4,3], 0.4*0.6**2*0.4, ""),
    ([3,2,3,2,3], 0.4*0.6*0.4*0.6, ""),
    ([3,2,3,2,1], 0.4*0.6*0.4**2, ""),
    ([3,2,1,2,3], 0.4**2*0.6**2, ""),
    ([3,2,1,2,1], 0.4**2*0.6*0.4, ""),
    ([3,2,1,0], 0.4**3, "→ $0 RUINA en n=3 (no continúa)"),
]

print(f"\nTotal de caminos: {len(caminos)}\n")

for i, (camino, prob, nota) in enumerate(caminos, 1):
    camino_str = ' → '.join([f'${x}' for x in camino])
    nota_str = f"  {nota}" if nota else ""
    print(f"{i:2}. {camino_str:30} | P = {prob:.6f} ({prob*100:6.3f}%){nota_str}")

# Fortuna esperada
fortuna_esperada = sum(camino[-1] * prob for camino, prob, _ in caminos)
print("\n" + "="*70)
print(f"Fortuna esperada en n=4: E[X₄] = ${fortuna_esperada:.2f}")
print("="*70)
```
