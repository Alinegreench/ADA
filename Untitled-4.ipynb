```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos
IQ = [70,74,78,82,86,90,94,98,102,106,110,114,118,122,126]
f = [4,9,16,28,45,66,85,72,54,38,27,18,11,5,2]

# Crear DataFrame
df = pd.DataFrame({'IQ': IQ, 'f': f})

# Total de alumnos
N = sum(f)

# Frecuencia relativa
df['fr'] = df['f']/N

# Frecuencia acumulada
df['F'] = df['f'].cumsum()
df['FR'] = df['fr'].cumsum()

print("Tabla de frecuencias:")
print(df)

# Media
media = sum(df['IQ'] * df['f']) / N
print("\nMedia:", round(media,2))

# Desviación típica
varianza = sum(df['f'] * (df['IQ'] - media)**2) / N
desviacion_tipica = np.sqrt(varianza)
print("Desviación típica:", round(desviacion_tipica,2))

# Desviación media
desviacion_media = sum(df['f'] * abs(df['IQ'] - media)) / N
print("Desviación media:", round(desviacion_media,2))

# Rango semi-intercuartil
Q1 = np.percentile(np.repeat(IQ, f), 25)
Q3 = np.percentile(np.repeat(IQ, f), 75)
RSI = (Q3 - Q1)/2
print("Rango semi-intercuartil:", round(RSI,2))

# Rangos percentiles
P10 = np.percentile(np.repeat(IQ, f), 10)
P90 = np.percentile(np.repeat(IQ, f), 90)
Rango_10_90 = P90 - P10
print("Rango percentil 10-90:", round(Rango_10_90,2))

P20 = np.percentile(np.repeat(IQ, f), 20)
P80 = np.percentile(np.repeat(IQ, f), 80)
Rango_20_80 = P80 - P20
print("Rango percentil 20-80:", round(Rango_20_80,2))

# Histograma de frecuencias relativas
plt.bar(df['IQ'], df['fr'], width=4, color='skyblue', edgecolor='black')
plt.xlabel('IQ')
plt.ylabel('Frecuencia relativa')
plt.title('Histograma de frecuencias relativas')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de la tabla
iq_valores = [70, 74, 78, 82, 86, 90, 94, 98, 102, 106, 110, 114, 118, 122, 126]
frecuencia_absoluta = [4, 9, 16, 28, 45, 66, 85, 72, 54, 38, 27, 18, 11, 5, 2]
frecuencia_relativa = [0.0083333, 0.01875, 0.0333333, 0.0583333, 0.09375, 
                      0.1375, 0.1770833, 0.15, 0.1125, 0.0791667, 
                      0.05625, 0.0375, 0.0229167, 0.0104167, 0.0041667]

# Verificar que las frecuencias relativas suman 1
print(f"Suma de frecuencias relativas: {sum(frecuencia_relativa):.6f}")

# Crear el histograma
plt.figure(figsize=(10, 6))
plt.bar(iq_valores, frecuencia_relativa, width=3, color='skyblue', 
        edgecolor='black', alpha=0.7)

# Personalizar el gráfico
plt.title('Histograma de Frecuencias Relativas - IQ', fontsize=16, pad=20)
plt.xlabel('IQ', fontsize=12)
plt.ylabel('Frecuencia Relativa', fontsize=12)
plt.grid(True, alpha=0.3)

# Configurar los ejes
plt.xlim(65, 130)
plt.ylim(0, 0.20)  # Máximo cerca de 0.18, no 1.75

# Añadir líneas de referencia
plt.axhline(y=0.1, color='red', linestyle='--', alpha=0.5, label='0.1')
plt.axhline(y=0.15, color='red', linestyle='--', alpha=0.5, label='0.15')

# Mostrar valores exactos en las barras más altas
for i, (x, y) in enumerate(zip(iq_valores, frecuencia_relativa)):
    if y > 0.1:  # Solo mostrar valores altos
        plt.text(x, y + 0.005, f'{y:.3f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

# Alternativa con histograma más tradicional usando bins
plt.figure(figsize=(10, 6))

# Crear datos expandidos para usar hist() de matplotlib
datos_expandidos = []
for iq, freq in zip(iq_valores, frecuencia_absoluta):
    datos_expandidos.extend([iq] * freq)

# Crear bins centrados en los valores de IQ
bins = np.arange(68, 130, 4)  # Bins de 68 a 128 con ancho 4

plt.hist(datos_expandidos, bins=bins, density=True, color='lightblue', 
         edgecolor='black', alpha=0.7)

plt.title('Histograma de Frecuencias Relativas - IQ', fontsize=16)
plt.xlabel('IQ', fontsize=12)
plt.ylabel('Frecuencia Relativa', fontsize=12)
plt.grid(True, alpha=0.3)
plt.xlim(65, 130)

plt.tight_layout()
plt.show()

# Mostrar estadísticas básicas
print("\nEstadísticas de los datos:")
print(f"Total de observaciones: {sum(frecuencia_absoluta)}")
print(f"IQ promedio ponderado: {sum(iq * freq for iq, freq in zip(iq_valores, frecuencia_absoluta)) / sum(frecuencia_absoluta):.2f}")
print(f"Valor máximo de frecuencia relativa: {max(frecuencia_relativa):.6f}")
print(f"IQ con mayor frecuencia: {iq_valores[frecuencia_relativa.index(max(frecuencia_relativa))]}")
```
