```python
# Datos del problema
procesos = [212, 417, 112, 426]       # Tamaño de cada proceso (MB)
bloques = [100, 500, 200, 300, 600]   # Tamaño de cada bloque de memoria (MB)
```

```python
# Estrategia: Primer bloque disponible
def primer_bloque_disponible(procesos, bloques):
    asignaciones = [-1] * len(procesos)  # Guarda qué bloque tiene cada proceso
    bloques_restantes = bloques[:]        # Copia para no modificar la original

    for i, proceso in enumerate(procesos):
        for j, bloque in enumerate(bloques_restantes):
            if proceso <= bloque:  # Si el bloque tiene suficiente espacio
                asignaciones[i] = j
                bloques_restantes[j] -= proceso  # Reducir el espacio disponible
                break  # Pasar al siguiente proceso

    # Mostrar resultados
    for i, proceso in enumerate(procesos):
        if asignaciones[i] != -1:
            print(f"Proceso {i+1} ({proceso} MB) → Bloque {asignaciones[i]+1}")
        else:
            print(f"Proceso {i+1} ({proceso} MB) → No asignado")
    print("Bloques restantes:", bloques_restantes)
    print()
```

```python
primer_bloque_disponible(procesos, bloques)
```

```python
# Estrategia: Mejor bloque disponible
def mejor_bloque_disponible(procesos, bloques):
    asignaciones = [-1] * len(procesos)
    bloques_restantes = bloques[:]

    for i, proceso in enumerate(procesos):
        mejor_idx = -1
        mejor_tamaño = float('inf')
        for j, bloque in enumerate(bloques_restantes):
            if proceso <= bloque and bloque < mejor_tamaño:
                mejor_tamaño = bloque
                mejor_idx = j
        if mejor_idx != -1:
            asignaciones[i] = mejor_idx
            bloques_restantes[mejor_idx] -= proceso

    for i, proceso in enumerate(procesos):
        if asignaciones[i] != -1:
            print(f"Proceso {i+1} ({proceso} MB) → Bloque {asignaciones[i]+1}")
        else:
            print(f"Proceso {i+1} ({proceso} MB) → No asignado")
    print("Bloques restantes:", bloques_restantes)
    print()
```

```python
mejor_bloque_disponible(procesos, bloques)
```
